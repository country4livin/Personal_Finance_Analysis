```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('Resources/Personal_Finance_Dataset.csv', parse_dates=['Date'])

# Inspect data
print(df.head())
print(df.info())
print(df['Category'].value_counts())
```

```python
# Add Month column
df['Month'] = df['Date'].dt.to_period('M')
```

```python
# Filter expenses only
expenses = df[df['Type'] == 'Expense']
```

```python
# Monthly spending
monthly_spending = expenses.groupby('Month')['Amount'].sum()
monthly_spending.plot(kind='bar', title='Monthly Spending')
plt.ylabel('Amount ($)')
plt.show()
```

```python
# Category-level breakdown
category_spending = expenses.groupby('Category')['Amount'].sum().sort_values(ascending=False)
category_spending.plot(kind='barh', title='Spending by Category')
plt.xlabel('Amount ($)')
plt.show()
```

```python
#Build budget optimizer
monthly_income = df[(df['Type'] == 'Income')]['Amount'].sum() / df['Month'].nunique()
recommended_budget = {
    'Needs (50%)': monthly_income * 0.5,
    'Wants (30%)': monthly_income * 0.3,
    'Savings/Debt (20%)': monthly_income * 0.2,
}
print("Recommended Monthly Budget:")
for k, v in recommended_budget.items():
    print(f"{k}: ${v:.2f}")
```
